Attemption 1:$O(n^2)$

```python
class Solution(object):
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """
        t=col(board)+row(board)+block(board)
        if t==0:
            return True
        else:
            return False

def block(board):
    a=[0]*10
    for i in range(3):
        for j in range(3):
            for t in range(3):
                for q in range(3):
                    if board[3*i+t][3*j+q]=='.':
                        continue
                    a[int(board[3*i+t][3*j+q])]=a[int(board[3*i+t][3*j+q])]+1;
                    if a[int(board[3*i+t][j*3+q])]==2:
                        print(a)
                        return 1
            for m in range(10):
                a[m]=0
    return 0
    

def row(board):
    a=[0]*10
    for i in range(9):
        for j in range(9):
            if board[i][j]=='.':
                continue
            a[int(board[i][j])]=a[int(board[i][j])]+1;
            if a[int(board[i][j])]==2:
                return 1
        for t in range(10):
            a[t]=0
    return 0

def col(board):
    a=[0]*10
    for j in range(9):
        for i in range(9):
            if board[i][j]=='.':
                continue
            a[int(board[i][j])]=a[int(board[i][j])]+1;
            if a[int(board[i][j])]==2:
                return 1
        for t in range(10):
            a[t]=0
    return 0
```
