# Leetcode question1:

给定一个长度为 n 的整数数组 height 。有 n 条垂线，第 i 条线的两个端点是 (i, 0) 和 (i, height[i]) 。

找出其中的两条线，使得它们与 x 轴共同构成的容器可以容纳最多的水。

返回容器可以储存的最大水量。

说明：你不能倾斜容器。

Attemption 1:
Out of time.
O(n^2)

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        count=len(height)
        maxn=0
        for i in range(count-1):
            alist=range(count)
            for j in alist[i+1:count]:
                cur=min( height[i],height[j] )*(j-i)
                if cur>maxn:
                    maxn=cur
        return maxn
```

Attemption 2:O(n)

```python
class Solution(object):
    def maxArea(self, height):
        le=0
        ri=len(height)-1
        maxwater=0
        curwater=0
        while ri>=le+1:
            curwater=(ri-le)*min(height[ri],height[le])
            if curwater>maxwater:
                maxwater=curwater
            if height[ri]>height[le]:
                le=le+1
            else:
                ri=ri-1
        return maxwater
```
